```python
to_remove = ['objectid', 'mapc_id', 'muni_id', 'parloc_id', 'poly_typ', 'map_num', 'mappar_id', 'loc_id_cnt', 'luc_1', 'luc_2', 'luc_adj_1', 'luc_adj_2', 'addr_num', 'addr_zip', 'owner_name', 'owner_addr', 'owner_city', 'owner_zip', 'pct_imperv', 'pct_bldg', 'pct_pave', 'bldlnd_rat', 'sqm_imperv', 'temp', 'shape']
```

```python
len(to_remove)
```

```python
re_mass = df.drop(to_remove, axis=1)
```

```python
high_corr = []
for a in range(len(re_mass.corr().columns)):
    for b in range(a):
        if abs(re_mass.corr().iloc[a,b]) >0.75:
            p = re_mass.corr().columns[a]
#            print(p)
            high_corr.append(p)
```
