```python
plt.figure(figsize=(12,8))
sns.countplot(y='Town', data=rs, order=rs['Town'].value_counts().index[:10])
plt.title('Number of Sales by Town (Top 10)')
plt.xlabel('Number of Sales')
plt.ylabel('Town')
plt.show()
```

```python
# Sale Amount vs. Assessed Value by Property Type
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Assessed Value', y='Sale Amount', hue='Property Type', data=rs, alpha=0.5)
plt.title('Sale Amount vs. Assessed Value by Property Type')
plt.xlabel('Assessed Value ($)')
plt.ylabel('Sale Amount ($)')
plt.legend(title='Property Type')
plt.show()
```

```python
# Sales Amount Trend Over Time
plt.figure(figsize=(10, 6))
rs.groupby('List Year')['Sale Amount'].sum().plot(marker='o')
plt.title('Total Sales Amount Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales Amount ($)')
plt.grid(True)
plt.show()
```

```python
# Top 10 Towns with Highest Average Sale Amount
top_towns_avg_sale = rs.groupby('Town')['Sale Amount'].mean().nlargest(10)

plt.figure(figsize=(12, 8))
top_towns_avg_sale.plot(kind='bar', color='skyblue')
plt.title('Top 10 Towns with Highest Average Sale Amount')
plt.xlabel('Town')
plt.ylabel('Average Sale Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
import geopandas as gpd
import plotly.express as px

# Load the geographical data (replace 'geodata.geojson' with your file path)
geodata = gpd.read_file('/content/drive/MyDrive/dataset/Real Estate Sales 2001-2021 GL_20240425.geojson')

# Merge rs DataFrame with geodata based on a common key
merged_data = pd.merge(geodata, rs, how='left', left_on='Town', right_on='Town')

# Create the choropleth map
fig = px.choropleth_mapbox(
    merged_data,
    geojson=merged_data.geometry,
    locations=merged_data.index,
    color='Assessed Value',  # Adjust this to the column you want to visualize
    mapbox_style='carto-positron',
    center={'lat': merged_data.geometry.centroid.y.mean(), 'lon': merged_data.geometry.centroid.x.mean()},
    zoom=8,
    opacity=0.7,
    hover_name='Town',  # Adjust this to the column you want to display on hover
    labels={'Assessed Value': 'Assessed Value'}  # Adjust label if needed
)

fig.show()
```
